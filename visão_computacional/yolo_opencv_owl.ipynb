```python
import cv2
import numpy as np
from rdflib import Graph, URIRef, RDF, RDFS, OWL, Literal
import matplotlib.pyplot as plt
from PIL import Image

# Função para carregar classes do arquivo coco.names
def load_coco_classes(file_path):
    with open(file_path, 'r') as file:
        return [line.strip() for line in file.readlines()]

# Carregar classes do COCO dataset
coco_classes = load_coco_classes('coco.names')

# Construir a ontologia (seu código RDF aqui)

# Configurar o modelo YOLO (substituir com seus próprios caminhos)
yolo_net = cv2.dnn.readNet("yolov3.weights", "yolov3.cfg")

# Modificar classes para usar instâncias das classes DeEspecifico, DeGenerico e Categoria
classes_yolo = coco_classes
layer_names = yolo_net.getUnconnectedOutLayersNames()

# Carregar a imagem
image = cv2.imread("girafa.jpg")

# Verificar se a imagem foi carregada com sucesso
if image is None:
    print("Erro ao carregar a imagem.")
    exit()

# Obter as dimensões da imagem
height, width, _ = image.shape

# Converter a imagem para o formato blob
blob = cv2.dnn.blobFromImage(image, 0.00392, (416, 416), (0, 0, 0), True, crop=False)

# Configurar a entrada da rede neural
yolo_net.setInput(blob)

# Obter as informações de saída da rede
outs = yolo_net.forward(layer_names)

# Lista para armazenar as classes detectadas, confianças e caixas delimitadoras
class_ids = []
confidences = []
boxes = []

# Loop sobre as detecções
for out in outs:
    for detection in out:
        scores = detection[5:]
        class_id = np.argmax(scores)
        confidence = scores[class_id]

        if confidence > 0.5 and 0 <= class_id < len(classes_yolo):
            # Obter as coordenadas da caixa delimitadora
            center_x = int(detection[0] * width)
            center_y = int(detection[1] * height)
            w = int(detection[2] * width)
            h = int(detection[3] * height)

            # Calcular as coordenadas do canto superior esquerdo da caixa delimitadora
            x = int(center_x - w / 2)
            y = int(center_y - h / 2)

            # Adicionar informações à lista
            class_ids.append(class_id)
            confidences.append(float(confidence))
            boxes.append([x, y, w, h])

# Aplicar supressão não máxima para remover detecções redundantes
indices = cv2.dnn.NMSBoxes(boxes, confidences, 0.5, 0.4)

# Mapear cores para cada classe
class_colors = {
    "car": (255, 0, 0),      # Vermelho
    "person": (0, 255, 0),   # Verde
    "bicycle": (0, 0, 255),  # Azul
    "bus": (255, 255, 0)     # Amarelo
}

# Processar os resultados da detecção
for i in indices:
    i = i[0]
    if 0 <= i < len(boxes):
        box = boxes[i]
        x, y, w, h = box
        label_yolo = classes_yolo[class_ids[i]]
        confidence = confidences[i]

        # Desenhar a caixa delimitadora na imagem com a cor correspondente
        color = class_colors.get(label_yolo, (0, 0, 0))  # Preto se a classe não estiver no mapeamento
        cv2.rectangle(image, (x, y), (x + w, y + h), color, 2)

        # Adicionar o nome da classe e confiança à caixa delimitadora
        label = f"{label_yolo}: {confidence:.2f}"
        cv2.putText(image, label, (x, y - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)

# Exibir a imagem com as caixas delimitadoras no Jupyter Notebook
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.axis('off')
plt.show()

# Imprimir as inferências
for i in indices:
    i = i[0]
    if 0 <= i < len(boxes):
        box = boxes[i]
        label_yolo = classes_yolo[class_ids[i]]
        confidence = confidences[i]

        # Criar inferência usando a ontologia
        inference = f"{label_yolo} detectado com confiança {confidence:.2f}"

        # Mapear a classe YOLO para as classes Categoria, DeGenerico e DeEspecifico na ontologia
        if label_yolo == "car":
            inference += f" - Categoria: oque - DeGenerico: Veiculo - DeEspecifico: Carro"
        elif label_yolo == "person":
            inference += f" - Categoria: quem - DeGenerico: SerBio - DeEspecifico: Pessoa"
        elif label_yolo == "bicycle":
            inference += f" - Categoria: oque - DeGenerico: Veiculo - DeEspecifico: Bicicleta"
        elif label_yolo == "bus":
            inference += f" - Categoria: oque - DeGenerico: Veiculo - DeEspecifico: Onibus"
        elif label_yolo == "giraffe":
            inference += f" - Categoria: quem - DeGenerico: SerBio - DeEspecifico: girafa"

        # Imprimir a inferência
        print(inference)
```
